```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

# Método Gráfico

### Restricciones

Generar un arreglo de x1 lineales para evaluar cada restricción

```python
x1 = np.linspace(0,4,10)
```

```python
# x1+x2 >= 2
r1_x2=2-x1
# -x1+x2 >= 1
r2_x2=1+x1
# x2 <= 3
r3_x2=3+0*x1
```

### Función de costo

Se debe calcular la función de costo en terminos de las cotas para establecer el mínimo.

```python
def calcularCota(z):
    return (x1/2)+z

#Cota -1
x2_N1=calcularCota(-1)
#Cota 0
x2_0=calcularCota(0)
#Cota 1
x2_1=calcularCota(1)
#Cota 2
x2_2=calcularCota(2)
#Cota 3
x2_3=calcularCota(3)
```

```python
plt.figure(figsize=(8,5))
plt.plot(x1,r1_x2,label='x1+x2>=2')
plt.plot(x1,r2_x2,label='-x1+x2>=1')
plt.plot(x1,r3_x2,label='x2<=3')
plt.text(2.7,0.4,'f(x)=-1',rotation=15,fontsize=12)
plt.plot(x1,x2_N1,'k:', linewidth=0.8)
plt.text(2,1.1,'f(x)=0',rotation=15,fontsize=12)
plt.plot(x1,x2_0,'k:', linewidth=0.8)
plt.text(1.5,1.9,'f(x)=1',rotation=15,fontsize=12)
plt.plot(x1,x2_1,'k:', linewidth=0.8)
plt.text(0.7,2.5,'f(x)=2',rotation=15,fontsize=12)
plt.plot(x1,x2_2,'k:', linewidth=0.8)
plt.text(0.2,3.2,'f(x)=3',rotation=15,fontsize=12)
plt.plot(x1,x2_3,'k:', linewidth=0.8)
plt.plot(0.5,1.5,'ro',label='f(x*)')
plt.xlim((0,3.5))
plt.ylim((0,3.5))
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend(loc='upper right')
#plt.grid()
plt.show();
```
