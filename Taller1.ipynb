```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import seaborn as sns
import sys
%matplotlib inline
```

# Método Gráfico

### Función de costo

f(x)=x1-2x2

```python
fig = plt.figure(figsize=(10,8))
ax = plt.axes(projection='3d')
x1,x2= np.meshgrid(np.linspace(0,4), np.linspace(0,4))
z=x1-2*x2
ax.plot_surface(x1,x2,z,cmap='Spectral_r',rstride=1, cstride=1)
ax.set_xlabel('x1')
ax.set_ylabel('x2')
ax.set_zlabel('f(x)')
ax.view_init(60,35);
```

```python
plt.contourf(x1,x2,z,10,cmap='Spectral');
plt.xlabel('x1')
plt.ylabel('x2');
```

### Restricciones

Se debe calcular la función de costo en terminos de las cotas para graficar las curvas de nivel

```python
x1_l = np.linspace(0,4,10)
```

```python
def calcularCota(z):
    return (x1_l/2)+z

#Cota -1
x2_N1=calcularCota(-1)
#Cota 0
x2_0=calcularCota(0)
#Cota 1
x2_1=calcularCota(1)
#Cota 2
x2_2=calcularCota(2)
#Cota 3
x2_3=calcularCota(3)
```

```python
# x1+x2 >= 2
r1_x2=2-x1_l
# -x1+x2 >= 1
r2_x2=1+x1_l
# x2 <= 3
r3_x2=3+0*x1_l
```

```python
plt.figure(figsize=(8,5))
plt.plot(x1_l,r1_x2,label='x1+x2>=2')
plt.plot(x1_l,r2_x2,label='-x1+x2>=1')
plt.plot(x1_l,r3_x2,label='x2<=3')
plt.text(2.7,0.4,'f(x)=-1',rotation=15,fontsize=12)
plt.plot(x1_l,x2_N1,'k:', linewidth=0.8)
plt.text(2,1.1,'f(x)=0',rotation=15,fontsize=12)
plt.plot(x1_l,x2_0,'k:', linewidth=0.8)
plt.text(1.5,1.9,'f(x)=1',rotation=15,fontsize=12)
plt.plot(x1_l,x2_1,'k:', linewidth=0.8)
plt.text(0.7,2.5,'f(x)=2',rotation=15,fontsize=12)
plt.plot(x1_l,x2_2,'k:', linewidth=0.8)
plt.text(0.2,3.2,'f(x)=3',rotation=15,fontsize=12)
plt.plot(x1_l,x2_3,'k:', linewidth=0.8)
plt.plot(0.5,1.5,'ro',label='f(x*)')
plt.xlim((0,3.5))
plt.ylim((0,3.5))
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend(loc='upper right')
#plt.grid()
plt.show();
```
